```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df = pd.read_csv(r'C:\Users\uday\Desktop\python\Task-2 data.csv')
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Fare'] = titanic_df['Fare'].fillna(titanic_df['Fare'].median())
titanic_df['Embarkation point'] = titanic_df['Embarkation point'].fillna(titanic_df['Embarkation point'].mode()[0])
titanic_df.drop(columns=['Cabin'], inplace=True)
titanic_df['Status'] = titanic_df['Status'].map({0: 'Not Survived', 1: 'Survived'})
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Status', hue='Sex', data=titanic_df)
plt.title('Survival Count by Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=titanic_df)
plt.title('Fare by Pclass')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Status', hue='Embarkation point', data=titanic_df)
plt.title('Survival Count by Embarkation Point')
plt.show()
```

```python
numeric_df = titanic_df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
