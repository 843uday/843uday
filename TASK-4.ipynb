```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
twitts_train = pd.read_csv(r'C:\Users\uday\Desktop\python\twitter_training.csv')
twitts_valid = pd.read_csv(r'C:\Users\uday\Desktop\python\twitter_validation.csv')
column_name = ['TweetID', 'Entity', 'Sentiment', 'Content']
twitts_train.columns = column_name
twitts_valid.columns = column_name
twitts = pd.concat([twitts_train, twitts_valid], ignore_index=False)
twitts.dropna(inplace=True)
twitts.drop_duplicates(inplace=True)
twitts.drop(columns=['TweetID', 'Content'], inplace=True)
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=twitts, y='Entity', order=twitts['Entity'].value_counts().index)
plt.title('Distribution of Entities')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
```

```python
entity_counts = twitts['Entity'].value_counts()
plt.figure(figsize=(18, 12))
entity_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Entities')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=twitts, x='Sentiment', order=twitts['Sentiment'].value_counts().index)
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
entity_sentiment = pd.crosstab(twitts['Entity'], twitts['Sentiment'])
plt.figure(figsize=(12, 8))
sns.heatmap(entity_sentiment, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Entity vs Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Entity')
plt.show()
```
